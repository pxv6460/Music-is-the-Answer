```python
x_axis = input('Value for X axis?\n')
y_axis = input('Value for Y axis?\n')
third_dimension = input('Value for third dimension?\n')
```

```python
x = music.filter([x_axis])
y = music.filter([y_axis])
w = music.filter([third_dimension])
```

```python
plt.scatter(x , y , w,  marker = 'o', facecolors = 'red', edgecolors = 'white')
plt.show()
```

```python
plt.style.use('_mpl-gallery')

xs = x
ys = y
zs = w

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()

```
